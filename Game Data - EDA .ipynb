```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel("03_Data.xlsx")
df.head(5)
```

```python
df.isnull().sum()
```

```python
df['active_days'] = df['active_days'].fillna(0)
df['inactive for days'] = df['inactive for days'].fillna(0)
df['sessions'] = df['sessions'].fillna(0)
df['session_length'] = df['session_length'].fillna(0)
df['total_playtime'] = df['total_playtime'].fillna(0)
df['ch_fights'] = df['ch_fights'].fillna(0)
df['fs_fights'] = df['fs_fights'].fillna(0)
df['ta_fights'] = df['ta_fights'].fillna(0)
df['last_fight_index'] = df['last_fight_index'].fillna(0)
df['robots_owned'] = df['robots_owned'].fillna(0)
df = df.drop(['ltv'], axis=1)
```

```python
# Do not touch
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

# 2. Univariate analysis - 3 Examples

```python
plt.figure(figsize=(10, 8))
sns.histplot(y='last_ch_fightname', data=df)
plt.ylabel( "Last flight" , size = 14 )
plt.xlabel( "Count" , size = 14 )
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="active_days", data=df)
plt.xlabel( "Active days" , size = 14 )
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="last_fight_index", data=df)
plt.xlabel( "last fight index" , size = 14 )
```

# 3. Bivariate analysis - 2 Examples

```python
plt.figure(figsize=(10, 6))
plt.title('Sessions & active days' , size = 14 )
plt.ylabel( "sessions" , size = 14 )
plt.xlabel( "active days" , size = 14 )
sns.lineplot(data=df, x="active_days", y="sessions")
```

```python
plt.figure(figsize=(10, 6))
plt.title('Session length & active days' , size = 14 )
plt.ylabel( "session length" , size = 14 )
plt.xlabel( "Active day " , size = 14 )
sns.lineplot(data=df, x="active_days", y="session_length")
```

# 4.	Visualize the following 
a.	Robots owned vs. Total playtime

b.	Game mode vs. robots owned


```python
plt.figure(figsize=(10, 6))
plt.title( "Total playtime vs robots owned" , size = 14 )
plt.xlabel( "Robots owned" , size = 14 )
plt.ylabel( "Total playtime" , size = 14 )
sns.lineplot(data=df, x="robots_owned", y="total_playtime")
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="robots_owned", y="ta_fights")
plt.title( "Arcade fight mode vs robots owned" , size = 14 )
plt.ylabel( "No of fights in time arcade fight mode" , size = 14 )
plt.xlabel( "Robots owned" , size = 14 )
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="robots_owned", y="ch_fights")
plt.title( "Championship fight mode vs robots owned" , size = 14 )
plt.ylabel( "No of fights in championship fight mode" , size = 14 )
plt.xlabel( "Robots owned" , size = 14 )
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="robots_owned", y="fs_fights")
plt.title( "Sparring fight mode vs robots owned" , size = 14 )
plt.ylabel( "No of fights in free sparring fight mode" , size = 14 )
plt.xlabel( "Robots owned" , size = 14 )
```
